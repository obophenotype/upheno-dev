```python
import os
import pandas as pd

def handle_gocc_anatomical_entity(group, df_obsoletes, dir_patterns):
    upheno_id_cc = None
    upheno_id_ae = None
    pattern = None

    for index, row in group.iterrows():
        print(row)
        if "AnatomicalEntity" in row["subject_id"] and "GO_" in row["subject_id"]:
            if upheno_id_ae is None:
                upheno_id_ae = row["Subject"]
            else:
                print(f"AE already set.. {row['Subject']} {row['subject_id']} {row['Subject']}")
        elif "CellularComponent" in row["subject_id"] and "GO_" in row["subject_id"]:
            if upheno_id_cc is None:
                upheno_id_cc = row["Subject"]
            else:
                print(f"CC already set.. {row['Subject']} {row['subject_id']} {row['Subject']}")
        else:
            print(f"Unexpected subject_id {row['subject_id']} {row['Subject']}")

        if pattern is None and "AnatomicalEntity" in row["subject_id"]:
            pattern = row["subject_id"].split("-")[-1]

    if upheno_id_cc is not None and upheno_id_ae is not None and pattern is not None:
        new_obsolete_entry = pd.DataFrame([{
            "replacement": upheno_id_cc,
            "obsolete": upheno_id_ae,
            "comment": f"The {upheno_id_cc} was redundant with {upheno_id_ae}, just using a different pattern."
        }])
        
        df_obsoletes = pd.concat([df_obsoletes, new_obsolete_entry], ignore_index=True)


        pattern_file = os.path.join(dir_patterns, pattern)
        if os.path.exists(pattern_file):
            df_pattern = pd.read_csv(pattern_file)
            print(df_pattern.head())
            print(df_pattern.columns)
            if "defined_class" in df_pattern.columns:
                print(df_pattern["defined_class"])
                df_pattern = df_pattern[df_pattern["defined_class"] != upheno_id_ae]
                df_pattern.to_csv(pattern_file, index=False)
            else:
                print(df_pattern)

    return df_obsoletes

grouped_df = df_robot_report_merged[df_robot_report["Rule Name"] == "duplicate_definition"].groupby("Value")

all_cases = []
upheno_ids = [
    "UPHENO:0050409",
    "UPHENO:0002001",
    "UPHENO:0002002",
    "UPHENO:0002003",
    "UPHENO:0002004",
    "UPHENO:0050590",
    "UPHENO:0051262"
]

print(f"Group count{len(grouped_df)}")

i = 0
for definition, group in grouped_df:
    i = i + 1
    new_case = True
    replacements = []
    go_bp = False
    go_mf = False
    case_type = "unknown"
    for index, row in group.iterrows():
        if "MolecularFunction" in row["subject_id"]:
            go_mf = True
        if "BiologicalProcess" in row["subject_id"]:
            go_bp = True
        
        ### Fixing all cases
        if "AnatomicalEntity" in row["subject_id"] and "GO_" in row["subject_id"]:
            # Obsolete the term, replace with other (CC), discard from matches
            new_case = False
            case_type = "gocc_anatomical_entity"
            break
        elif "brokenAnatomicalEntity" in row["subject_id"] and "UBERON_" in row["subject_id"]:
            # Fixed by itself (changed pattern text)
            new_case = False
            case_type = "broken_fractured"
            break
        elif "LevelOfChemicalEntity" in row["subject_id"] and "PR_" in row["subject_id"]:
            # Just one fix manually
            new_case = False
            case_type = "pr_chemical_entity"
            break
        elif "CellularComponent" in row["subject_id"] and "CL_" in row["subject_id"]:
            # Big issue with CL being both CC and AE.
            # Fix:
            # 1. Drop CL from CC patterns entirely, and replace with CL-AE term
            # 2. Obsolete CL from CC terms
            # 3. Add check to matches to not include CL-CC cases
            new_case = False
            case_type = "cc_cl"
            break
        elif "abnormalFunctionalityOfPartOfAnatomicalEntity" in row["subject_id"]:
            # Fixed by itself (changed pattern text)
            new_case = False
            case_type = "functionality_partof"
            break
        elif "BiologicalProcess" in row["subject_id"] and "NBO_" in row["subject_id"]:
            # Smaller issue with NBO terms overlapping with BP.
            # Fix:
            # 1. Drop NBO from BP patterns entirely, and replace with BP term
            # 2. Obsolete NBO from BP terms
            # 3. Add check to matches to not include NBO-BP cases
            # 4. Output NBO-BP mappings to ensure NBO pattern matches are not excluded..
            new_case = False
            case_type = "nbo_bp"
            break
        elif "ResistanceOfWholeOrganismToEntity" in row["subject_id"] and "CHEBI_" in row["subject_id"]:
            # Biggest issue with CHEBI being both "entity" and "chemical entity".
            # Fix:
            # 1. Drop CHEBI from ResistanceOfWholeOrganismToEntity patterns entirely, and replace with CHEBI-ResistanceOfWholeOrganismToChemicalEntity term
            # 2. Obsolete CHEBI-ENTITY
            # 3. Add check to matches to not include CHEBI-ENTITY cases
            new_case = False
            case_type = "resistence_to_entity_chebi"
            break
        elif any(upheno_id in row["Subject"] for upheno_id in upheno_ids):
            # Already fixed
            new_case = False
            case_type = "fixed_id"
            break
        elif go_bp and go_mf:
            # All of these are molecular function. Delete the BP ones, add replacements
            new_case = False
            case_type = "bp_and_mf"
            break
        
    if new_case:
        print(f"Case {i}")
        print(group)
        break
    else:
        if case_type == "gocc_anatomical_entity":
            handle_gocc_anatomical_entity(group, df_obsoletes, automatic_data_dir)

# Create a DataFrame from the global list
#df_replacements = pd.DataFrame(all_cases)

#df_replacements.to_csv("replacements.tsv", sep="\t", index=False)
#df_replacements.value_counts("case_type")
df_obsoletes
```
